## Questão 9

```python
for i in range(eventos.shape[0]):
    evento = eventos.loc[i,'evento']
    inicio = eventos.loc[i,'data_inicial']
    #Somando um dia para que o intervalo seja até menor que o inicio do dia posterior
    fim = eventos.loc[i,'data_final'] + timedelta(days=1)
    duracao = fim - inicio
    chamados_eventos = chamados_2.loc[(chamados_2['data_inicio'] >= inicio) & (chamados_2['data_inicio'] < fim),:]
    nchamados = chamados_eventos.shape[0]
    chamadosevento = [evento,nchamados/duracao.days]
    print(chamadosevento)
```

#### Resposta:
| Evento | Média Diária |
| ----------- | ----------- |
| Carnaval | 60.25 |
| Reveillon | 45.66... |
| Rock in Rio (1) | 122.0 |
| Rock in Rio (2)| 117.0 |
